```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
df= pd.read_csv('cars.csv')
df
```

Using Pandas


```python
pd.get_dummies(df, columns=['fuel','owner'], drop_first='True')
```
